# TESS data

```python
import numpy as np
import matplotlib.pyplot as plt
import os
from astropy.io import fits
from astropy.table import Table
```

```python
p1 = os.getcwd()
p2 = p1 + '/Data/TESS/'
```

```python
lcf = fits.open(p2 + 'tess2018206045859-s0001-0000000149603524-0120-s_lc.fits')
lcf_h = lcf[0].header
lcf_h
```

```python
lcf_d = Table.read(lcf[1])
```

```python
lcf_d
```

```python
tp = fits.open(p2 + 'tess2018206045859-s0001-0000000149603524-0120-s_tp.fits')
tp_h = tp[0].header
tp_h
```

```python
tp_d = Table.read(tp[1])
tp_d
```

```python
plt.figure(figsize=(16/1.5, 9/1.5))
plt.errorbar(tp_d['POS_CORR1'], lcf_d['PDCSAP_FLUX'], fmt='.')
```

```python
pos1, fl = np.asarray(lcf_d['POS_CORR1']), np.asarray(lcf_d['PDCSAP_FLUX'])
fl, pos1 = fl[np.isfinite(fl)], pos1[np.isfinite(fl)]
plt.figure(figsize=(16/1.5, 9/1.5))
plt.errorbar(pos1, fl, fmt='.')
plt.xlim(-0.05,0.04)
```

```python
dvt = fits.open(p2 + 'tess2018206190142-s0001-s0001-0000000149603524-00106_dvt.fits')
dvt_h = dvt[0].header
dvt_h
```

```python
dvt_d = dvt[1].data
dvt_d
```

```python
plt.figure(figsize=(16/1.5,9/1.5))
plt.errorbar(dvt_d['TIME'],dvt_d['LC_DETREND'],fmt='.')
```
